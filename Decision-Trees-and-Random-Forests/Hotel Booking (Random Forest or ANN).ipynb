```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("hotel_bookings.csv")
df.dtypes
```

```python

# dropping column having too many NULL values
df.drop(["company"],axis=1,inplace=True)
df=df.dropna()
print("Columns : NULL values")
# deleting the rows having null values
for i in df.columns:
    print(i,df[i].isna().sum())

# reset the index values
df=df.reset_index(drop=True)
df
```

```python
df["reservation_status_date"] = pd.to_datetime(df.reservation_status_date)
df["reservation_status_day"]=list(df.reservation_status_date.dt.weekday)
df["reservation_status_month"]=list(df.reservation_status_date.dt.month)
df["reservation_status_week"]=list(df.reservation_status_date.dt.week)
df["reservation_status_year"]=list(df.reservation_status_date.dt.year)
df.drop(["reservation_status_date"],axis=1,inplace=True)
df
```

```python
month={}
for i in range(0,12):#list(df["arrival_date_month"].unique()):
    if i<6:
        month[list(df["arrival_date_month"].unique())[i]] = i+7
    else:
        month[list(df["arrival_date_month"].unique())[i]] = i-5 
for i in range(0,df.shape[0]):
    df["arrival_date_month"].values[i]=month[df["arrival_date_month"].values[i]]
```

```python
month
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["arrival_date_year"]=label_encoder.fit_transform(df["arrival_date_year"].values)
df["reservation_status_year"]=label_encoder.fit_transform(df["reservation_status_year"].values)
df["hotel"]=label_encoder.fit_transform(df["hotel"].values)
df["meal"]=label_encoder.fit_transform(df["meal"].values)
df["country"]=label_encoder.fit_transform(df["country"].values)
df["market_segment"]=label_encoder.fit_transform(df["market_segment"].values)
df["distribution_channel"]=label_encoder.fit_transform(df["distribution_channel"].values)
df["reserved_room_type"]=label_encoder.fit_transform(df["assigned_room_type"].values)
df["deposit_type"]=label_encoder.fit_transform(df["deposit_type"].values)
df["customer_type"]=label_encoder.fit_transform(df["customer_type"].values)
df["reservation_status"]=label_encoder.fit_transform(df["reservation_status"].values)
df["assigned_room_type"]=label_encoder.fit_transform(df["assigned_room_type"].values)
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
    print("\n")
```

```python
X = df.drop("is_canceled",axis=1)
y = df["is_canceled"]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=0)
```

#  Random Forest Test

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
rfc = RandomForestClassifier(n_estimators = 40)
rfc.fit(X_train,y_train)
rfc_pred = rfc.predict(X_test)
print(classification_report(y_test,rfc_pred))
print('\n')
print(confusion_matrix(y_test,rfc_pred))
```
